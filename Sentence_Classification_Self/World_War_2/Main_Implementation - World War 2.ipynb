```python
from CNN import tCNN
```

### Getting Data

```python
import pickle
with open('Data_worldwar2.pkl', 'rb') as f:
    Final = pickle.load(f)
    
X = Final[0]
Y = Final[1]
```

### Random Shuffle

```python
import random

XY = list(zip(X, Y))
random.shuffle(XY)

X, Y = zip(*XY)
```

```python
tCNN()


from keras.utils import plot_model
from IPython.display import Image
from keras.models import load_model

model = load_model('CNN_model_init.h5')
plot_model(model, to_file='CNN_arch.png')
Image(filename='./CNN_arch.png')
```

## Reshaping array

```python
import numpy as np

X = np.asarray(X)
Y = np.asarray(Y)

X = X.reshape(X.shape[0], X.shape[1], X.shape[2], 1)
Y = Y.reshape(Y.shape[0], 1)

print(f"X shape : {X.shape}\nY shape : {Y.shape}")
```

```python
x_train = X[0:140]
y_train = Y[0:140]

x_test = X[140:]
y_test = Y[140:]
```

### Training Model

```python
# callback function
import keras
import matplotlib.pyplot as plt

class myCallback(keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs={}):
        if(logs.get('acc')>0.90):
            print("\nReached 90% accuracy so cancelling training!")
            self.model.stop_training = True
        
callback = myCallback()

history = model.fit(x_train, y_train,
          epochs=1000,
          verbose=1,
          validation_data=(x_test, y_test),
          callbacks=[callback])

try:
    model.save('./CNN_model_worldwar2.h5')
    print("Model Saved!")
except:
    print("Error in saving Model")
```

## Plotting Model Accuracy and Loss

```python
# Plot training & validation accuracy values
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.savefig('Model_accuracy_worldwar2.png', bbox_inches='tight')
plt.show()

# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.savefig('Model_loss_worldwar2.png', bbox_inches='tight')
```

```python
score = model.evaluate(x_test, y_test, verbose=0)
score
```
