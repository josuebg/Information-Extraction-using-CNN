```python
from LineSplit import LineSplit
```

```python
# check files

files = ['summary_worldwar2.txt', 'Ref_WorldWar2_chandni.txt', 'Ref_WorldWar2_mihir.txt', 'Ref_WorldWar2_sweta.txt']

for file in files:
    print(f"Lines Count {file}: ", len(LineSplit(file)))
```

```python
common_matrix = []

for fOne in files:
    listOne = LineSplit(fOne)
    row = []
    
    for fTwo in files:
        listTwo = LineSplit(fTwo)
        cLines = []
        for line in listOne:
            if line in listTwo:
                cLines.append(line)
                listTwo.remove(line)
        row.append(len(cLines))
    common_matrix.append(row)
```

```python
common_matrix
```

```python
import seaborn as sn
import pandas as pd
import matplotlib.pyplot as plt

df_cm = pd.DataFrame(common_matrix, index = ['candidate', 'ref_chandni', 'ref_mihir', 'ref_sweta'],
                  columns = ['candidate', 'ref_chandni', 'ref_mihir', 'ref_sweta'])


sn.set(rc={'figure.figsize':(15.7,10.27)})
sn.set(font_scale=1.4)#for label size
sn.heatmap(df_cm, annot=True,annot_kws={"size": 16})# font size
plt.title(label="Common Lines Confusion Matrix of World War 2 Classifier", pad=10.0)
plt.savefig('confusion_commonLines_worldwar2.png')
plt.show()
```
